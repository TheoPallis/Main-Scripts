```python
import os
import glob

# This will search for .ipynb files in the current directory and all subdirectories.
# Change the search_path to a specific directory if you want to search a different location
search_path =  "C:/**/*.ipynb"

# List to hold all matching files
matching_files = []

# Walk through the directory
for filename in glob.glob(search_path, recursive=True):
    # Check if the file name (without extension) is more than 4 characters
    if len(os.path.splitext(os.path.basename(filename))[0]) > 4:
        matching_files.append(filename)

# Print out the list of matching .ipynb files
print("IPython Notebooks with more than 4 characters in their name:")
for file in matching_files:
    print(file)
```
